**단위 원뿔(반지름 20, 높이 30)의 부피와 겉넓이를 출력하는 문제**

```python
# 원뿔 계산 함수 정의
def prt_cone_vol_surf(r, h) :
    if r > 0 and h > 0 :
        # r, h 모두 양수일 때
        sla = pow(r ** 2 + h ** 2, 0.5)
        return 1/3 * 3.14 * r ** 2 * h, 3.14 * r ** 2 + 3.14 * r * sla
    else :
        # r, h가 음수일 때
        return -1, -1
```
